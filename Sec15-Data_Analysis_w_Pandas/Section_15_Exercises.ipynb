```python
########################################################
# Add Lat/Long to Pandas data frame
########################################################
import pandas
df=pandas.read_csv("supermarkets.csv")
```

```python
df["FullAddress"] = df["Address"] +", "+ df["City"] +", "+ df["State"] +", "+ df["Country"]
df
```

```python
df["Coordinates"] = df["FullAddress"].apply(nom.geocode)
df
```

```python
df.Coordinates[0]
```

```python
df["latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```
